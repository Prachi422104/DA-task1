```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
#this shows the first 5 rows 
df.head()
```

```python
#this shows the content whether the any data is missing or not , inshort counts and gives us information about all columns and rows
df.info()
```

```python
#this counts the nukl enteries of each column
df.isnull().sum()
```

```python
#the director column has lot of missing values we cannot fill them so we'll replace it by not available
df['director'] = df['director'].fillna('Not Available')
```

```python
#the cast column is same as director we cannot put by ourselve so we'll replace it with not available
df['cast'] = df['cast'].fillna('Not Available')
```

```python
#the country column is same as above replacing it with unknown 
df['country'] = df['country'].fillna('Unknown')
```

```python
#lets check what changes has done till now
df.info()
```

```python
df.isnull().sum()
```

```python
#handling the date 
#since date is not available and only 10 rows are missing we can drop them 
df = df.dropna(subset=['date_added'])
```

```python
#handling the rating, filling the missing values using mode
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
#handling the duration only 3 missing values lets fill with unknown
df['duration'] = df['duration'].fillna('Unknown')
```

```python
#lets check again
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
#saving the cleaned file 
df.to_csv("netflix_cleaned.csv", index=False)
```
